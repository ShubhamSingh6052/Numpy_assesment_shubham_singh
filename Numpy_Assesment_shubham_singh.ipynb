```python
#1).Import pandas and read in the banklist.csv file into a dataframe called banks.

import pandas as pd
banks = pd.read_csv('banklist.csv')
banks
```

```python
#2). Show the head of the dataframe.

print(banks.head())
```

```python
#3).What are the column names?

column_names = banks.columns
print("Column names:", column_names)
```

```python
#4).How many States (ST) are represented in this data set?

# Get the number of unique states (ST) represented in the data set
num_states = banks['ST'].nunique()
print("Number of states represented:", num_states)
```

```python
#5).Get a list or array of all the states in the data set.

states_list = banks['ST'].unique()
print("List of states:", states_list)
```

```python
#6).What are the top 5 states with the most failed banks?

top_5_states = banks['ST'].value_counts().head(5)
print("Top 5 states with the most failed banks:\n", top_5_states)
```

```python
#7).What are the top 5 acquiring institutions?

top_5_acquiring_institutions = banks['Acquiring Institution'].value_counts().head(5)
print("Top 5 acquiring institutions:\n", top_5_acquiring_institutions)
```

```python
#8).How many banks has the State Bank of Texas acquired? How many of them were actually in Texas?

state_bank_of_texas_acquisitions = banks[banks['Acquiring Institution'] == 'State Bank of Texas'].shape[0]
print("Number of banks acquired by the State Bank of Texas:", state_bank_of_texas_acquisitions)

state_bank_of_texas_in_texas = banks[(banks['Acquiring Institution'] == 'State Bank of Texas') & (banks['ST'] == 'TX')].shape[0]
print("Number of banks acquired by the State Bank of Texas that were in Texas:", state_bank_of_texas_in_texas)
```

```python
#9).What is the most common city in California for a bank to fail in?

most_common_city_in_CA = banks[banks['ST'] == 'CA']['City'].value_counts().idxmax()
print("Most common city in California for a bank to fail in:", most_common_city_in_CA)
```
